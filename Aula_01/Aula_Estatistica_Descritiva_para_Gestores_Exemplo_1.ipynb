```python
#Lendo os dados
dados = pd.read_excel("Tabela 6407.xlsx")
```

```python
#Limpando espaços vazios e linha 0 (Total)
dados = dados.drop(0).dropna()
```

```python
dados
```

```python
#Descrição estatística básica
dados.describe()
```

```python
#Cálculo de frequência em porcentagem
dados["Freq (%)"] = dados["Total"] * 100 / dados["Total"].sum()
```

```python
#Cálculo de frequência acumulada
dados["Freq Acumulada"] = dados["Freq (%)"].cumsum()
```

```python
dados
```

```python
#Descartando 60 anos ou mais
dados = dados.drop(10)
```

```python
dados
```

```python
#Recalculando frequências com a exclusão de 60 anos ou mais
dados["Freq (%)"] = dados["Total"] * 100 / dados["Total"].sum()
dados["Freq Acumulada"] = dados["Freq (%)"].cumsum()
```

```python
#Criando coluna com os dados da idade mediana de cada trecho
dados["Idade mediana"] = [2, 9, 15.5, 18.5, 22, 27, 34.5, 44.5, 54.5]
```

```python
#Cálculo da média considerando a idade mediana dos intervalos
(dados["Idade mediana"] * dados["Total"]).sum() / dados["Total"].sum()
```
